## Clustering x-ray data

```python
from keras.preprocessing.image import (
    array_to_img,
    img_to_array,
    load_img,
    ImageDataGenerator,
)
from tqdm.notebook import tqdm
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

# First we will work with the Montgomery X-Ray set. Plot all

```python
montgomery_path = "../../data/MontgomerySet/CXR_png/*.png"
xray_files = glob(montgomery_path, recursive=False)
fig, axes = plt.subplots(12, 12, figsize=(20, 20))
flat_axes = axes.flat
idx = 0
for idx, xray in enumerate(xray_files):
    ax = flat_axes[idx]
    xray_im = cv2.imread(xray)
    xray_im = cv2.resize(xray_im, (50, 50))
    ax.imshow(cv2.cvtColor(xray_im, cv2.COLOR_BGR2RGB))
    ax.axis('off')

for jdx in range(idx, len(flat_axes)):
    ax = flat_axes[jdx]
    ax.axis('off')
```

```python
fig, axes = plt.subplots(12, 12, figsize=(20, 20))
flat_axes = axes.flat
for idx, ax in enumerate(flat_axes):
    ax.text(0, 0, f"{idx}", fontsize=42, color='b')
    ax.axis('off')
```
